```python
childrenData = pd.read_csv(r'C:\Users\danny\Downloads\UNdata_Export_20230715_051224289.csv')
childrenData
```

```python
childrenData['Number of children ever-born alive'].unique()
```

```python
#Cleaning because a lot of cells in Number of children living are in string or nonnumeric
numChild = childrenData[(childrenData.get('Area') == 'Total') & (childrenData.get('Age') == '15 +') 
                        & (childrenData.get('Number of children ever-born alive') != 'Total') 
                        & (childrenData.get('Number of children ever-born alive') != 'Not stated') 
                        & (childrenData.get('Number of children ever-born alive') != 'Aggregate number of children ever-born')]
# numChild['Number of children ever-born alive'] = numChild['Number of children ever-born alive'].replace('Not stated', np.nan)
numChild['Number of children ever-born alive'] = pd.to_numeric(numChild['Number of children ever-born alive'], errors='coerce')
numChild = numChild.dropna(subset=['Number of children ever-born alive'])
numChild
```

```python
# Querying, generating the statistic
numChild = numChild[['Country or Area', 'Year', 'Number of children ever-born alive', 'Value']] 
numChild = numChild.rename(columns = {'Value': 'Number of Women'})
numChild['Number of Children'] = [i * j for i, j in zip(numChild['Number of children ever-born alive'], numChild['Number of Women'])] 
numChild = numChild.drop(columns = 'Number of children ever-born alive')
numChild = numChild.groupby(by=['Country or Area', 'Year']).sum()
numChild = numChild.reset_index()

numChild['Number of Children Ever Born per Woman'] = [i / j for i, j in zip(numChild['Number of Children'], numChild['Number of Women'])] 
numChild
```
