```python
import numpy as np
import pandas as pd
```

```python
childrenData = pd.read_csv(r'C:\Users\danny\OneDrive\Files\GitHub\research-lab-task2\Data\Number of children living per woman.csv')
```

```python
childrenData
```

```python
#Cleaning because a lot of cells in Number of children living are in string or nonnumeric
numChild = childrenData[(childrenData.get('Area') == 'Total') & (childrenData.get('Age') == '15 +') 
                        & (childrenData.get('Number of children living') != 'Total') & (childrenData.get('Number of children living') != 'Aggregate number of children living')]
numChild['Number of children living'] = numChild['Number of children living'].replace('Not stated', np.nan)
numChild['Number of children living'] = pd.to_numeric(numChild['Number of children living'], errors='coerce')
numChild = numChild.dropna(subset=['Number of children living'])

numChild
```

```python
# Querying, generating the statistic
numChild = numChild[['Country or Area', 'Year', 'Number of children living', 'Value']] 
numChild = numChild.rename(columns = {'Value': 'Number of Women'})
numChild['Number of Children'] = [i * j for i, j in zip(numChild['Number of children living'], numChild['Number of Women'])] 
numChild = numChild.drop(columns = 'Number of children living')
numChild = numChild.groupby(by=['Country or Area', 'Year']).sum()
numChild = numChild.reset_index()

numChild['Number of Children Living per Woman'] = [i / j for i, j in zip(numChild['Number of Children'], numChild['Number of Women'])] 
numChild
```

```python
numChild = numChild[['Country or Area', 'Year', 'Number of Children Living per Woman']]
numChild
```

```python
numChild.to_csv('Number of children living per woman.csv', index=False)
```
