```python
import numpy as np
import pandas as pd
```

```python
LFP_by_sex = pd.read_csv(r'C:\Users\danny\OneDrive\Files\GitHub\research-lab-task2\Views\Labor Force Participation by Sex.csv')
LFP_by_sex_household = pd.read_csv(r'C:\Users\danny\OneDrive\Files\GitHub\research-lab-task2\Views\Labor force participation, by sex, of households with kids 6 and under.csv')
childrenBorn = pd.read_csv(r'C:\Users\danny\OneDrive\Files\GitHub\research-lab-task2\Views\Number of Children Ever Born per Woman.csv')
childrenLiving = pd.read_csv(r'C:\Users\danny\OneDrive\Files\GitHub\research-lab-task2\Views\Number of children living per woman.csv')

allOther = pd.read_csv(r'C:\Users\danny\OneDrive\Files\GitHub\research-lab-task2\Views\Report on education, literacy and employment.csv')
```

```python
allOther = allOther.drop(columns='Unnamed: 0')
allOther
```

```python
# LFPcombine = LFP_by_sex.merge(LFP_by_sex_household, left_on=['Country or Area', 'Year'], right_on= ['Country or Area', 'Year'])
LFPcombine = LFP_by_sex.merge(LFP_by_sex_household, how='outer')
LFPcombine
```

```python
# childrenCombine = childrenBorn.merge(childrenLiving, left_on=['Country or Area', 'Year'], right_on= ['Country or Area', 'Year'])
childrenCombine = childrenBorn.merge(childrenLiving, how='outer')
childrenCombine
```

```python
# combine = LFPcombine.merge(childrenCombine, left_on=['Country or Area', 'Year'], right_on= ['Country or Area', 'Year'])
combine = LFPcombine.merge(childrenCombine, how='outer')
combine
```

```python
final = combine.merge(allOther, how='outer')
final
```

```python
final.to_csv('FinalResult.csv', index=False)
```
