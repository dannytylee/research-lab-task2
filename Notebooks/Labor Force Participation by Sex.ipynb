```python
populationData = pd.read_csv(r'C:\Users\danny\Downloads\UNdata_Export_20230721_003833507.csv')
populationData
```

```python
populationData['Sex'].unique()
```

```python
populationData = populationData[(populationData.get('Sex') != 'Both Sexes') & (populationData.get('Age') == 'Total') & (populationData.get('Area') == 'Total')]
totalPopulation = populationData[(populationData.get('Activity status') == 'Total population')]
activePopulation = populationData[(populationData.get('Activity status') == 'Total economically active population')]
```

```python
totalPopulation = totalPopulation[['Country or Area', 'Year', 'Sex', 'Value']]
totalPopulation = totalPopulation.rename(columns = {'Value': 'Total population'})
totalPopulation
```

```python
activePopulation = activePopulation[['Country or Area', 'Year', 'Sex', 'Value']]
activePopulation = activePopulation.rename(columns = {'Value': 'Total economically active population'})
activePopulation
```

```python
result = activePopulation.merge(totalPopulation, left_on= ['Country or Area', 'Year', 'Sex'], right_on=['Country or Area', 'Year', 'Sex'])
result['Labor force participation'] = [i / j * 100 for i, j in zip(result['Total economically active population'], result['Total population'])] 
result
```

```python
maleResult = result[result.get('Sex') == 'Male']
maleResult
```

```python
femaleResult = result[result.get('Sex') == 'Female']
femaleResult
```
