### labor force participation, by sex, of households with kids 6 and under

```python
import numpy as np
import pandas as pd
```

```python
populationData = pd.read_csv(r'C:\Users\danny\Downloads\GED_2LFP_SEX_RT_A-full-2023-07-21.csv')
populationData
```

```python
populationData = populationData[['ref_area.label', 'sex.label', 'time', 'obs_value']]
maleData = populationData[populationData.get('sex.label').str.contains('Male')]
femaleData = populationData[populationData.get('sex.label').str.contains('Female')]
```

```python
maleData = maleData.rename(columns = {'ref_area.label': 'Country or Area', 'time': 'Year', 'sex.label': 'Sex', 'obs_value': 'Labor force participation'})
maleData['Sex'] = ['Male' for i in zip(maleData['Sex'])] 
maleData
```

```python
femaleData = femaleData.rename(columns = {'ref_area.label': 'Country or Area', 'time': 'Year', 'sex.label': 'Sex', 'obs_value': 'Labor force participation'})
femaleData['Sex'] = ['Female' for i in zip(femaleData['Sex'])] 
femaleData
```
